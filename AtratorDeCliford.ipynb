```python
"""
Diciplina: Física Computacional 
Autor: Carlos Euardo a Silva Lima
Data: 03/06/2021
Tema: Atrator e cliford
"""
#Apresentação
print("Atrator de Cliford")
#Importando bibliotecas matemáticas 
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('dark_background')
#
a = -2.1
b = -0.8
c = 1.2
d = 1.6
#  
N = 1000000
x = np.zeros(N)
y = np.zeros(N)
#
x[0] = 1
y[0] = 1
for i in range(N-1):
    x[i+1] = np.sin(a*y[i]) + c*np.cos(a*x[i])
    y[i+1] = np.sin(b*x[i]) + d*np.cos(b*y[i])
    # print("x[",i,"] = ",x[i]," || y[",i,"] = ",y[i])

# Fazer gráfico
plt.figure(figsize = (10,10))
plt.plot(x,y, '^', color='white', alpha = 2, markersize=0.3)
plt.axis('on')
plt.show()
plt.close()
```
