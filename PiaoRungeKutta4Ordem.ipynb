```python
'''
Autor: Carlos Eduardo da Silva Lima 
Disciplina: Física Computacional 
Trabalho: Trabalho de Física Computacional
Tema: Caos, vibrações mecânicas, Runge-Kutta de 4Ordem - Pênulo Duplo
'''
# Importação de Pacotes e bibliotecas
# from matplotlib.cbook import get_sample_data
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('dark_background')

# Atributos, parâmetros do pião
I1 = 1
I2 = 1
I3 = 1
m = 1
g = 9.81
l = 1

# Condições inicias 
t_0  = 0 # Tempo inicial
x1_0 = 0 # 
x2_0 = 0 # 
x3_0 = 0 # 
x4_0 = 0 # 

# Passo h e quantiade de iterações 
h = 0.01 # passo
N = 100  # Quantidade de Iterações

# Equações Diferenciais
# Primeira EDO
def f1(t,x1,x2,x3,x4):
  return (x2)

def f2(t,x1,x2,x3,x4):
  return (x3)

def f3(t,x1,x2,x3,x4):
  return (x4)

def f4(t,x1,x2,x3,x4):
  return ((np.power(x3,2)*np.sin(x1)*np.cos(x1)*(I1-I3) - I3*x3*x4*np.sin(x1) + m*g*l*np.sin(x1))/I1)

def f5(t,x1,x2,x3,x4):
  return ((2*(I3-I1)*x2*x3*np.sin(x1)*np.cos(x1) - I3*x2*x3*np.sin(x1)*np.cos(x1) + I3*x2*x4*np.sin(x1))/I1*np.power(np.sin(x1),2))

def f6(t,x1,x2,x3,x4):
  return ((np.cos(x1)/I1*np.power(np.sin(x1),2))*(((I1*np.power(np.sin(x1),2) + I3*np.power(np.cos(x1),2))/np.cos(x1))*x2*x3*np.sin(x1) - 2*(I3-I1)*x2*x3np.sin(x1)*np.cos(x1) - I3*x2*x4*np.sin(x1)))


# Arrays, variáveis para a inplementação no métoo de Runge-Kutta de 4Ordem
t  = np.empty(N + 1)
x1 = np.empty(N + 1)
x2 = np.empty(N + 1)
x3 = np.empty(N + 1)
x4 = np.empty(N + 1)

# Inicialização
t[0]  = t_0
x1[0] = x1_0
x2[0] = x2_0
x3[0] = x3_0
x4[0] = x4_0

for i in range(N-1):

  k11 = h*f1(t[i],x1[i],x2[i],x3[i],x4[i])
  k12 = h*f2(t[i],x1[i],x2[i],x3[i],x4[i])
  k13 = h*f3(t[i],x1[i],x2[i],x3[i],x4[i])
  k14 = h*f4(t[i],x1[i],x2[i],x3[i],x4[i])
  k15 = h*f5(t[i],x1[i],x2[i],x3[i],x4[i])
  k16 = h*f6(t[i],x1[i],x2[i],x3[i],x4[i])

  k21 = h*f1(t[i]+(h/2),x1[i]+(k11/2),x2[i]+(k12/2),x3[i]+(k13/2),x4[i]+(k14/2))
  k22 = h*f2(t[i]+(h/2),x1[i]+(k11/2),x2[i]+(k12/2),x3[i]+(k13/2),x4[i]+(k14/2))
  k23 = h*f3(t[i]+(h/2),x1[i]+(k11/2),x2[i]+(k12/2),x3[i]+(k13/2),x4[i]+(k14/2))
  k24 = h*f4(t[i]+(h/2),x1[i]+(k11/2),x2[i]+(k12/2),x3[i]+(k13/2),x4[i]+(k14/2))
  k25 = h*f5(t[i]+(h/2),x1[i]+(k11/2),x2[i]+(k12/2),x3[i]+(k13/2),x4[i]+(k14/2))
  k26 = h*f6(t[i]+(h/2),x1[i]+(k11/2),x2[i]+(k12/2),x3[i]+(k13/2),x4[i]+(k14/2))

  k31 = h*f1(t[i]+(h/2),x1[i]+(k21/2),x2[i]+(k22/2),x3[i]+(k23/2),x4[i]+(k24/2))
  k32 = h*f2(t[i]+(h/2),x1[i]+(k21/2),x2[i]+(k22/2),x3[i]+(k23/2),x4[i]+(k24/2))
  k33 = h*f3(t[i]+(h/2),x1[i]+(k21/2),x2[i]+(k22/2),x3[i]+(k23/2),x4[i]+(k24/2))
  k34 = h*f4(t[i]+(h/2),x1[i]+(k21/2),x2[i]+(k22/2),x3[i]+(k23/2),x4[i]+(k24/2))
  k35 = h*f5(t[i]+(h/2),x1[i]+(k21/2),x2[i]+(k22/2),x3[i]+(k23/2),x4[i]+(k24/2))
  k36 = h*f6(t[i]+(h/2),x1[i]+(k21/2),x2[i]+(k22/2),x3[i]+(k23/2),x4[i]+(k24/2))

  k41 = h*f1(t[i]+h,x1[i]+k31,x2[i]+k32,x3[i]+k33,x4[i]+k34)
  k42 = h*f2(t[i]+h,x1[i]+k31,x2[i]+k32,x3[i]+k33,x4[i]+k34)
  k43 = h*f3(t[i]+h,x1[i]+k31,x2[i]+k32,x3[i]+k33,x4[i]+k34)
  k44 = h*f4(t[i]+h,x1[i]+k31,x2[i]+k32,x3[i]+k33,x4[i]+k34)
  k45 = h*f5(t[i]+h,x1[i]+k31,x2[i]+k32,x3[i]+k33,x4[i]+k34)
  k46 = h*f6(t[i]+h,x1[i]+k31,x2[i]+k32,x3[i]+k33,x4[i]+k34)

  x1[i+1] = x1[i] + ((k11 + 2*(k21 + k31) + k41)/6)
  x2[i+1] = x2[i] + ((k12 + 2*(k22 + k32) + k42)/6)
  x3[i+1] = x3[i] + ((k13 + 2*(k23 + k33) + k43)/6)
  x4[i+1] = x4[i] + ((k14 + 2*(k24 + k34) + k44)/6)
  t[i+1]  =  t[i] + h
  
# Fim do Loop For 
```
