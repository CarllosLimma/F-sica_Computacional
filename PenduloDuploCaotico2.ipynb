```python
#-------------------------------------------------------------------------------------------------------------------------------------------------------------------------
#                               Universidade Federal de Goiás (Ufg)
#
# Autor          : Carlos Eduardo da Silva Lima 
# Diciplina      : Física Computacional 
# Tema           : Pêndulo Duplo Caótico
# Método Nimérico: Runge-Kutta de 4Ordem 
# Data           : 24/07/2021
# Linguagem      : Python
# IDE            : Google Colaboratory
#-------------------------------------------------------------------------------------------------------------------------------------------------------------------------
#-------------------------------------------------------------------------------------------------------------------------------------------------------------------------
#                               Pacotes e Bibliotecas importadas

from matplotlib.cbook import get_sample_data
import matplotlib.pyplot as plt
import numpy as np

#-------------------------------------------------------------------------------------------------------------------------------------------------------------------------
plt.style.use('dark_background') # Estilo do gráfico - Dark
#-------------------------------------------------------------------------------------------------------------------------------------------------------------------------

#                               Atributos, parâmetros associado ao Pêndulo Duplo caótico

m1 = float(input("Entre com o valor de m1: ")) # Kg
m2 = float(input("Entre com o valor de m2: ")) # Kg
l1 = float(input("Entre com o valor de l1: ")) # m
l2 = float(input("Entre com o valor de l2: ")) # m
g  = float(input("Entre com o valor de g : ")) # m/s^2
#-------------------------------------------------------------------------------------------------------------------------------------------------------------------------

#                                       Condições iniciais

t_0     = float(input("Entre com o valor de     t01: "))             # s
teta1_0 = np.radians(float(input("Entre com o valor de teta1_0: "))) # radianos
teta2_0 = np.radians(float(input("Entre com o valor de teta2_0: "))) # radianos
w1_0    = float(input("Entre com o valor de    w1_0: "))             # radianos/s
w2_0    = float(input("Entre com o valor de    w2_0: "))             # radianos/s

#                 Parâmetros para o método de Runge-Kutta de 4Ordem (Olhe no algoritimo de Runge-Kutta de 4Ordem)

N = int(input("Entre com a quantidade de iterações N: "))
h = float(input("Entre com passo h                  : "))
#-------------------------------------------------------------------------------------------------------------------------------------------------------------------------
#                                         Equações Diferenciais de Primeira Ordem (Uma Sistema de EDO)

# Primeira EDO
def f1(t,x1,x2,v1,v2):
  return v1

# Segunda EDO
def f2(t,x1,x2,v1,v2):
  return v2

# Terceira EDO
def f3(t,x1,x2,v1,v2):
	return (-g*(2*m1+m2)*np.sin(x1)-m2*g*np.sin(x1-2*x2)-2*m2*np.sin(x1-x2)*(l2*np.power(v2,2)+l1*np.power(v1,2)*np.cos(x1-x2))/l1*(2*m1+m2-m2*np.cos(2*x1-2*x2)))

# Quarta EDO
def f4(t,x1,x2,v1,v2):
	return (2*np.sin(x1-x2)*(np.power(v1,2)*l1*(m1+m2)+g*(m1+m2)*np.cos(x1)+np.power(v2,2)*l2*m2*np.cos(x1-x2))/l2*(2*m1+m2-m2*np.cos(2*x1-2*x2)))

#-------------------------------------------------------------------------------------------------------------------------------------------------------------------------

#                                    Arrays, variáveis para a inplementação no métoo de Runge-Kutta de 4Ordem
t  = np.empty(N)
x1 = np.empty(N)
x2 = np.empty(N)
x3 = np.empty(N)
x4 = np.empty(N)

# Inicialização
t[0]  = t_0
x1[0] = teta1_0
x2[0] = teta2_0
x3[0] = w1_0
x4[0] = w2_0
#-------------------------------------------------------------------------------------------------------------------------------------------------------------------------
#                                                       Algoritimo de Runge-Kutta de 4Ordem
for i in range(N-1):
  
  k11 = h*f1(t[i],x1[i],x2[i],x3[i],x4[i])
  k12 = h*f2(t[i],x1[i],x2[i],x3[i],x4[i])
  k13 = h*f3(t[i],x1[i],x2[i],x3[i],x4[i])
  k14 = h*f4(t[i],x1[i],x2[i],x3[i],x4[i])

  k21 = h*f1(t[i]+(h/2),x1[i]+(k11/2),x2[i]+(k12/2),x3[i]+(k13/2),x4[i]+(k14/2))
  k22 = h*f2(t[i]+(h/2),x1[i]+(k11/2),x2[i]+(k12/2),x3[i]+(k13/2),x4[i]+(k14/2))
  k23 = h*f3(t[i]+(h/2),x1[i]+(k11/2),x2[i]+(k12/2),x3[i]+(k13/2),x4[i]+(k14/2))
  k24 = h*f4(t[i]+(h/2),x1[i]+(k11/2),x2[i]+(k12/2),x3[i]+(k13/2),x4[i]+(k14/2))

  k31 = h*f1(t[i]+(h/2),x1[i]+(k21/2),x2[i]+(k22/2),x3[i]+(k23/2),x4[i]+(k24/2))
  k32 = h*f2(t[i]+(h/2),x1[i]+(k21/2),x2[i]+(k22/2),x3[i]+(k23/2),x4[i]+(k24/2))
  k33 = h*f3(t[i]+(h/2),x1[i]+(k21/2),x2[i]+(k22/2),x3[i]+(k23/2),x4[i]+(k24/2))
  k34 = h*f4(t[i]+(h/2),x1[i]+(k21/2),x2[i]+(k22/2),x3[i]+(k23/2),x4[i]+(k24/2))

  k41 = h*f1(t[i]+h,x1[i]+k31,x2[i]+k32,x3[i]+k33,x4[i]+k34)
  k42 = h*f2(t[i]+h,x1[i]+k31,x2[i]+k32,x3[i]+k33,x4[i]+k34)
  k43 = h*f3(t[i]+h,x1[i]+k31,x2[i]+k32,x3[i]+k33,x4[i]+k34)
  k44 = h*f4(t[i]+h,x1[i]+k31,x2[i]+k32,x3[i]+k33,x4[i]+k34)

  x1[i+1] = x1[i] + ((k11 + 2*(k21 + k31) + k41)/6)
  x2[i+1] = x2[i] + ((k12 + 2*(k22 + k32) + k42)/6)
  x3[i+1] = x3[i] + ((k13 + 2*(k23 + k33) + k43)/6)
  x4[i+1] = x4[i] + ((k14 + 2*(k24 + k34) + k44)/6)
  t[i+1]  =  t[i] + h
  
# Fim do Loop For 


# Saida para as contas no loop anterior
for i in range(N-1):
  print("t = ",t[i]," || teta1 = ",x1[i]," || teta2 = ",x2[i]," || w1 = ",x3[i]," || w2 = ",x4[i])
# Fim do Loop For 


#-------------------------------------------------------------------------------------------------------------------------------------------------------------------------
#                                                         Plotagem de Gráficos


# Duas Figuras 
fig, (ax1, ax2) = plt.subplots(1, 2, constrained_layout=True, sharey=True)

ax1.plot(x1, x3,'r.') #ax1.plot(x1, x3,'r*')
ax1.set_title('teta1 x x1')
ax1.set_xlabel('teta1 (rad)')
ax1.set_ylabel('w1 (rad/s)')

ax2.plot(x2, x4,'b.') # ax2.plot(x2, x4,'b*')
ax2.set_title('teta2 x x2')
ax2.set_xlabel('teta2 (rad)')
ax2.set_ylabel('w2 (rad/s)')

fig.suptitle('Pêndulo Duplo Caótico', fontsize=16)

# Uma única figura
```
