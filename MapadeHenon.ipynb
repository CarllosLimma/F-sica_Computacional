```python
"""
Física Computacional UFG
Autor: Carlos Eduardo da Silva Lima
Tema: Mapa de Henon
Data: 03/06/2021
"""
#Importando bibliotecas matemáticas 
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('dark_background')
#  Parâmetros
a = 0.2
b = 0.9991
#  N: numeor de iterações , e variáveis arrays x e y de tamanho N
N = 1000000
x = np.zeros(N)
y = np.zeros(N)
# Condições iniciais para x e y
x[0] = 0
y[0] = 0
""" Dentro do loop for inplementamos o algoritmo para o mapoa de Henon
"""
for i in range(N-1):
    x[i+1] = 1 - a*x[i]*x[i] + y[i]
    y[i+1] = b*x[i]
    #print("x[",i,"] = ",x[i]," || y[",i,"] = ",y[i])

# Fazer gráfico
plt.figure(figsize = (10,10))
plt.plot(x,y, '^', color='white', alpha = 2, markersize=0.3)
plt.axis('on')
plt.show()
plt.close()
```
